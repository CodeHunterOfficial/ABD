Игра:

Можно создать викторину на основе этой базы знаний, где игрокам будут задаваться вопросы о свойствах и характеристиках различных животных. Вопросы могут быть представлены в виде множественного выбора или True/False. За каждый правильный ответ игрок получает очки, а в конце определяется победитель.

```python
import random

# База данных с информацией о животных
animals = {
    "Тюлень": {
        "Млекопитающие": True,
        "Птицы": False,
        "Рыбы": False,
        "Животные": True,
        "В воде": True
    },
    "Попугай": {
        "Млекопитающие": False,
        "Птицы": True,
        "Рыбы": False,
        "Животные": False,
        "В воде": False
    },
    "Тигр": {
        "Млекопитающие": True,
        "Птицы": False,
        "Рыбы": False,
        "Животные": True,
        "В воде": False
    },
    "Жираф": {
        "Млекопитающие": True,
        "Птицы": False,
        "Рыбы": False,
        "Животные": True,
        "В воде": False
    },
    "Коза": {
        "Млекопитающие": True,
        "Птицы": False,
        "Рыбы": False,
        "Животные": True,
        "В воде": False
    },
    "Синица": {
        "Млекопитающие": False,
        "Птицы": True,
        "Рыбы": False,
        "Животные": False,
        "В воде": False
    },
    "Антилопа": {
        "Млекопитающие": True,
        "Птицы": False,
        "Рыбы": False,
        "Животные": True,
        "В воде": False
    },
    "Кит": {
        "Млекопитающие": True,
        "Птицы": False,
        "Рыбы": True,
        "Животные": False,
        "В воде": True
    }
}

def play_quiz():
    score = 0
    animals_list = list(animals.keys())
    random.shuffle(animals_list)

    for animal in animals_list:
        print(f"Вопрос об животном: {animal}")

        for category in ["Млекопитающие", "Птицы", "Рыбы", "Животные", "В воде"]:
            answer = animals[animal][category]
            user_answer = input(f"Является ли {animal} {category.lower()}? (Да/Нет) ").lower() == "да"

            if user_answer == answer:
                print("Правильно!")
                score += 1
            else:
                print("Неправильно.")

        print()

    print(f"Ваш итоговый счет: {score}/{len(animals) * 5}")

play_quiz()
```
